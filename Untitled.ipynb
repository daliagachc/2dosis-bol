```python
rd = {
    'Santa Cruz de la Sierra':'SC',
    'Cochabamba':'Cbba'
}
```

```python
plt.rcParams.update({
  "text.usetex": True,
  "font.family": "Helvetica"
})
```

```python
np.random.seed(0)
d1[R] = np.random.randn(len(d1))
s1 = d1.sort_values(P,ascending=False).head(20)
s1[M] = s1[M].replace(rd)
Me = (d1[P]*d1[PR]).sum()/d1[P].sum()
cm = plt.get_cmap('tab10')
f,ax = plt.subplots(dpi=100,figsize=(9,5))

ax.set_xlim(-1,9)
ax.set_ylim(-0,111)
for nn in range(9):
# nn = 0
    _d = d1[d1[N]==nn].sort_values(P,ascending=False)


    ax.scatter(x=_d[R]/6+nn,y=_d[PR],s=_d[P]/1000,edgecolor='k',facecolor=cm(nn))
    
    me = (_d[P]*_d[PR]).sum()/_d[P].sum()
    
#     ax.scatter(nn,me,marker='*',c='r',s=100,edgecolor='k')
    
    ax.hlines(y=[me],xmin=nn-.5,xmax=nn+.5,color='.5',ls=':',zorder=-4)
    
    __d0 = _d.sort_values(PR,ascending=False)[:2]
    __d1 = _d.sort_values(PR,ascending=False)[-1:]
    
    _d2 = pd.concat([__d0,__d1])
    
    for i,r in _d2.iterrows():
        xx = r[NN]/600
        ax.text(r[N]+r[R]/6+xx/5,r[PR]+xx+1,r[M],c='.7',zorder=10,fontsize=r[NN]/500+5)
    
    
ax.text(8.6,me+1,'promedio\ndepartamental',va='center',fontsize='8',c='.8')
ax.set_xticks(list(dic.values()));
ax.set_xticklabels(list(dic.keys()),ha='right',rotation=30);

ax.axhline(Me,c='.8',ls='--',zorder=-5)

sns.despine(fig=f,trim=True)

ll = [100,90,85,81]
ax.scatter(y=np.array(ll)-5,x=[8.5]*4,s=np.array([1e6,1e5,1e4,1e3])/1000,c='.8')

ax.text(8.5,105,'    habitantes',ha='center',c='.7')

for i,t in zip(ll,[1e6,1e5,1e4,1e3]) :
    ax.text(y=i-5,x=8.9,s=f'{int(t):,}'.replace(',',','),c='.8',va='center',fontsize=f'{8+(t**.5)/5000}')
    
ax.text(9.1,Me+1,'promedio\nnacional',va='center',fontsize='9',c='.8')
ax.set_ylabel('Segunda dosis [\%]')
ax.set_xlabel('Departamento')


for i,r in s1.iterrows():
    xx = r[NN]/600
    ax.text(r[N]+r[R]/6+xx/5,r[PR]+xx+1,r[M],c='.7',zorder=10,fontsize=r[NN]/500+5)

    
for i,r in d1.iterrows():

    ax.text(r[N]+r[R]/6,r[PR],r[M],c='.7',zorder=10,fontsize=1)

f.savefig('dd.pdf')
```
